# Histograms of water parcel timing

```python
import xarray as xr
import datetime as dt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
tim = pd.read_csv('/ocean/rbeutel/MOAD/biogeo_paper/FRDR/model/ariane/summary_files/TimingData.csv')
```

```python
tim['days'] = (pd.to_timedelta(tim.age) / pd.Timedelta(days=1))
tim['days'] = tim["days"].astype(int)
tim
```

```python
# overall mean timing

fig, ax = plt.subplots(1,1, figsize = (4.5,4.5))

bins = np.arange(0,150,1)
ax.hist(tim.days, bins=bins, weights=tim.transport, linewidth = 1.6, histtype='stepfilled', zorder=2)

ax.set_xlabel('Days',fontsize=14)
ax.set_ylabel('Transport [$m^{3}s^{-1}$]',fontsize=14)
ax.set_xticks(np.arange(0,150,50))
ax.set_xticklabels(np.arange(0,150,50),fontsize=12)
ax.set_xlim([0,150])
ax.set_ylim([0,920000000])

# vertical line for 20 day buffer
ax.plot([20,20],[0,920000000],'k--')
```

## separating by water mass

```python
sections = ['cuc', 'off_d', 'off_s', 'north', 'south', 'fresh']
colours = ['#AA3377', '#4477AA','#66CCEE','#228833','#EE6677','#CCBB44']
labels =["CUC",'Offshore Deep', 'Offhsore Surface', 'North', 'South', 'South Brackish']
```

```python
days = [tim.days[tim.wm==sections[0]],tim.days[tim.wm==sections[1]],tim.days[tim.wm==sections[2]],tim.days[tim.wm==sections[3]],tim.days[tim.wm==sections[4]],tim.days[tim.wm==sections[5]]]
trans = [tim.transport[tim.wm==sections[0]],tim.transport[tim.wm==sections[1]],tim.transport[tim.wm==sections[2]],tim.transport[tim.wm==sections[3]],tim.transport[tim.wm==sections[4]],tim.transport[tim.wm==sections[5]]]

fig, ax = plt.subplots(1,1, figsize = (4.5,4.5))

bins = np.arange(0,150,1)

ax.hist(days, bins=bins, weights= trans, color = colours, label=labels , linewidth = 1.6, histtype='step', stacked=False, zorder=2)
ax.hist(days, bins=bins, weights=trans, color = colours, histtype='stepfilled', alpha=0.3, stacked=False, zorder=1)

ax.legend(fontsize=12)
ax.set_xlabel('Days',fontsize=14)
ax.set_ylabel('Transport [$m^{3}s^{-1}$]',fontsize=14)
ax.set_xticks(np.arange(0,150,50))
ax.set_xticklabels(np.arange(0,150,50),fontsize=12);
ax.set_xlim([0,150])
ax.set_ylim([0,140000000])


# vertical line for 20 day buffer
ax.plot([20,20],[0,920000000],'k--')

plt.tight_layout()
plt.savefig('/ocean/rbeutel/MOAD/biogeo_paper/PI_BIOGEO_PAPER/model/figures/timing.jpg')
```
